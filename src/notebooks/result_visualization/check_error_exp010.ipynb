```python
train_df["pred"] = np.nan
for i in range(5):
    pred = pd.read_csv(f"../experiments/exp_010/val_preds/preds_0_{i}.csv")
    train_df["pred"][train_df["timeSeries_cv_0"] == i] = pred["pred"].values
    print(pred["pred"].sum())
    print(pred.shape)
```
