```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline


y_ = pd.read_feather("../my_features/train/delayTime.feather")
pred = pd.read_csv("../experiments/exp_000/val_preds/oof_preds.csv")
plot_df = pd.concat([y_, pred], axis=1)
```

```python
plot_df = plot_df.value_counts(["delayTime", "pred"]).reset_index()
plot_df
```

```python
# generate data
x = plot_df["delayTime"].values
y = plot_df["pred"].values

d = np.linspace(0,50,50)

fig = plt.figure()

fig, ax = plt.subplots(figsize=(20, 10))

ax.scatter(x,y, s=2)
ax.plot(d, d,color='black')

ax.set_title('first scatter plot')
ax.set_xlabel('y')
ax.set_ylabel('pred')

fig.show()
```
