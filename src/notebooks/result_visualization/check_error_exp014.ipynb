```python
import numpy as np
import pandas as pd
import warnings

pd.set_option('display.max_rows', 1100)
pd.set_option('display.max_columns', 1100)
warnings.filterwarnings('ignore')

train_df = pd.read_csv("../experiments/exp_013/train/train.csv")
train_df["planArrival_int"] = train_df["planArrival"].map(lambda x : int(x.replace(":", "")))
train_df["pred"] = pd.read_csv("../experiments/exp_013/val_preds/oof_preds.csv").values
```

```python
train_df["max_time"] = train_df.groupby(["date", "trainNo"])["planArrival_int"].transform("max")
train_df["min_time"] = train_df.groupby(["date", "trainNo"])["planArrival_int"].transform("min")
```

```python
train_df.query("min_time <= 800").query("800 <= max_time")[300:500]
```
