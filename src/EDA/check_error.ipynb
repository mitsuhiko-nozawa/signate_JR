```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
train_df = pd.read_csv("../input/train.csv")
train_df["date * trainNo"] = train_df["date"].astype("str") + " * " + train_df["trainNo"]
train_df[["continuedDelayTime", "isnanDelayTime"]] = pd.read_feather("../my_features/train/continuedDelayTime.feather").values
train_df["pred"] = pd.read_csv("../experiments/exp_000/val_preds/oof_preds.csv").values
train_df["error"] = train_df.apply(lambda x : np.abs(x["delayTime"] - x["pred"]), axis=1)
```

```python
train_df.groupby("date * trainNo")["error"].mean().sort_values()[-100:]
```
